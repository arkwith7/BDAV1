```python
pid = 'NTN007'
format_1_years = ['2014', '2015', '2016', '2017', '2018']
format_2_years = ['2019', '2020', '2021']
usr_id = 'kjsjunsik11' 
key = 'BJRRDMV0U1BNYYJNA15E5O2JM6XY2X8C4WWZC60XJ3ARE'

df = pd.DataFrame()

for year in format_2_years:
    # year = '2019'
    print(year)
    url = f'https://www.recycling-info.or.kr/sds/JsonApi.do?PID={pid}&YEAR={year}&USRID={usr_id}&KEY={key}'

    # 요청 보내기
    response = requests.get(url)

    json_data = response.json()
    
    print('data load success')

    # 'data'를 DataFrame으로 변환
    data_df = pd.DataFrame(json_data['data'])
    # 컬럼명을 한글로 변환
    # dataHeader
    col_dict = json_data['dataHeader'][0]
    col_dict['WT_TYPE_GB_NM'] = '폐기물 종류_대'
    col_dict['WSTE_M_CODE_NM'] = '폐기물 종류_중'
    col_dict['WSTE_CODE_NM'] = '폐기물 종류_소'
    data_df.rename(columns = col_dict, inplace = True)

    # 데이터에 기준연도 추가
    data_df['YEAR'] = year

    df = pd.concat([df, data_df], ignore_index=True)
    
def clean_string(input_string):
    # \r을 _로 변환
    replaced_string = input_string.replace('\r', '_')
    # 문자열 앞뒤의 특수 문자 제거 (여기서는 모든 구두점을 제거 대상으로 설정)
    cleaned_string = replaced_string.strip(string.punctuation)
    return cleaned_string

df = df.applymap(lambda x: clean_string(x) if isinstance(x, str) else x)
```
